# Normalizacja
$$X=\{x_1,x_2,x_3,\ldots,x_n\} $$
$$z_i=\frac{xi-\min(X)}{\max(X)-\min(X)} $$ 


```python
min_temp = min(temp)
max_temp = max(temp)
diff=max_temp-min_temp
```

```python
max_temp
```

```python
min_temp
```

```python
min_temp_norm = [(xi-min_temp)/diff for xi in temp]
min_temp_norm
```

```python
min_temp_norm_x = range(len(min_temp_norm))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
fig.set_dpi(300)
ax.scatter(min_temp_norm_x, min_temp_norm, color="red")
ax.scatter(min_temp_norm_x, temp, color="blue")
ax.set_yticks(range(0,18,1))
ax.set_title("Normalizacja")
```

# Standaryzacja
$$ z_i=\frac{x_i-\bar{x}}{\sigma} $$
gdzie $\bar{x}$ to średnia zbioru $X$,

$\sigma$ to odchylenie standardowe zbioru $X$
Cechą charakerystyczną zbioru Z jest suma = 0

```python
temp_avg = np.mean(temp)
temp_avg
```

```python
temp_std = np.std(temp)
temp_std
```

```python
min_temp_stand = [(xi-temp_avg)/temp_std for xi in temp]
min_temp_stand
```

```python
np.sum(min_temp_stand) # suma nie jest równa zero przez dokładność związaną ze standardem IEE 754
```

```python
min_temp_stand_x = range(len(min_temp_stand))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(9, 5)
fig.set_dpi(500)
ax.scatter(min_temp_stand_x, min_temp_stand, color="red", label="standaryzacja")
ax.scatter(min_temp_stand_x, temp, color="blue", label="dane początkowe")
ax.set_yticks(range(0,18,1))
ax.legend()
```

$$ z_i = \frac{1}{1+exp(-x_i)} $$

```python
min_temp_sigm = [1/(1+np.e**(-xi)) for xi in temp]
min_temp_sigm
```

```python
min_temp_sigm_x = range(len(min_temp_sigm))
```

# Normalizacja sigmoidą 2
$$ z_i = \frac{e^{x_i}-e^{-x_i}}{e^{x_i}+e^{-x_i}} $$

```python
def sigm2(x):
    ex = np.e**x
    e_neg_x = np.e**(-x)
    return (ex-e_neg_x)/(ex+e_neg_x)
```

```python
min_temp_sigm2 = [sigm2(xi) for xi in temp]
min_temp_sigm2
```

```python
min_temp_sigm2_x = range(len(min_temp_sigm2))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
fig.set_dpi(500)
ax.scatter(min_temp_sigm2_x, min_temp_sigm2, color="red", label="normalizacja sigmoidą 2")
ax.scatter(min_temp_sigm2_x, temp, color="blue", label="dane początkowe")
ax.set_yticks(range(0,18,1))
ax.legend()
```
