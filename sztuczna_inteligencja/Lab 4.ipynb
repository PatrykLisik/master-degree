```python
pixel_value, pixel_count = np.unique(photo, return_counts=True)
```

```python
x_min = 0
x_max = 255
```

```python
def N0(t):
    suma = 0
    for i in range(x_min,t-1):
        suma += H.get(i,0)
    return suma
```

```python
def ms_phi(t):
    suma = 0
    for i in range(x_min,t-1):
        suma += i*H.get(i,0)
    return suma/N0(t)
```

```python
def p_phi(t):
    return N0(t)/N
```

```python
def P1(t):
    return 1-p_phi(t)
```

```python
def ms_1(t):
    suma = 0
    for i in range(t,x_max):
        suma += i*H.get(i,0)
    return suma/(N-N0(t))
```

```python
def to_find(t):
    return p_phi(t)*P1(t)*(ms_phi(t)-ms_1(t))**2
```

```python
vals = np.array([to_find(t) for t in range(2,255)])
```
