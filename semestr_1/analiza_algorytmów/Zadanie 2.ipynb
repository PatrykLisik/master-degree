```python
def n(x):
    return x
def log_2_n(x):
    return math.log2(x)
def n_log_2(x):
    return x*math.log2(x)
def factorial(n):
    return math.factorial(int(n))
def two_to_n(n):
    return 2**n

def flog2(x):
    """Floor of log2"""
    if x <= 0:
        raise ValueError("domain error")
    return x.bit_length() - 1

def log_2_n_fac(x):
    return flog2(math.gamma(int(x)))

def n_to_2(n):
    return n**2

def sqrt(n):
    return math.sqrt(n)


def sqrt_log_2_n(n): # i
    return math.sqrt(n)*math.log2(n)

def n_sqrt(n): # j
    return n*math.sqrt(n)

def log_2_n_squared(x): # k
    return math.log2(x)**2

def log_2_n_square(x): # l
    return 2*math.log2(x) 

def sqrt_n_log_2_n(x):
    return math.sqrt(x*math.log2(x))

def n_sqrt_log_2_n(x):
    return x*math.sqrt(math.log2(x))

def one(x):
    return 1

def n_log_n_square(x):
    return x*math.log2(x)**2
```

```python

def print_order(value, callables =all_functions):
    order ={}
    for func in callables:
#         print(f"Computing  {func.__name__}({value})")
        order[func(value)] = func.__name__
    for key in sorted(order):
        print(f"{key}-{order[key]}")
```

```python
@cache
def S(n):
    if n==0:
        return 0
    return S(np.floor(n/2))+n-2*np.floor(n/2)
```

```python
y=range(1,10**5)
x = [S(x1) for x1 in y]
fig, ax = plt.subplots()
fig.set_size_inches(20, 15)
ax.scatter(y, x, label="$S_n$")
plt.grid()
# ax.set_yscale("log") 
```

```python
for n in range(100):
    val = S(n)
    print(f"{n} = {val}")
```

```python
sum = 0
count = 10000000
vals= []
for n in range(count):
    val = S(n)
    sum +=val
    vals.append(val)
#     print(f"{n} = {val}")
print(f"sum: {sum} nlogn/2 = {(np.log2(n)*n)/2}")
```

```python
y=np.linspace(1,count)
grow = (np.log2(y))

fig, ax = plt.subplots()
fig.set_size_inches(25, 15)
ax.plot(y,grow, label="expected")
ax.scatter( range(count), vals)
ax.legend()
plt.grid()
```
