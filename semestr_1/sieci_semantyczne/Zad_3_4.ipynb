# Relacje notatka
symetryczna - zawsze aplikuje się w dwie strony (bycie kuzynem).

asymetrycna - nie może aplikować się w dwie strony (bycie rodzicem).

zwrotne - Właściwość wiąże element z samym sobą(posiadania krewnego)

przeciwzwortne - Właściwość nie może wiązać elementu z samym sobą(bycie rodzicem)

funcjonalne - dla danego elementu może przyjąć tylko jedną wartość(posiada PESEL)

odwrotnie funkcjonalne - dla dwóch różnych obiektów nie może posiadać tej samej wartości(posiadanie męża)

Właściwości przechodnie - podsiada przodka
