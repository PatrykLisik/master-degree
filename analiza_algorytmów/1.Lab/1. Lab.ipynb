### 16. Znajdź górne i dolne ograniczenie dla $\sum^n_{k=1}\sqrt{k}$

1. to nie jest szereg geometryczny

$p \le \floor{\sqrt{k}} \le p+1 $


$p+1 \le \ceil{\sqrt{k}} \le p $

raczej nie da się tego użyć 


```python
x = np.arange(1,50)
y = np.ceil(np.sqrt(x))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
fig.set_dpi(300)
ax.scatter(x, y)
ax.set_ylabel(r'$\lceil\sqrt{k}\rceil$')
ax.set_xlabel('$k$')
plt.grid()
```

```python
x = np.arange(1,5000)
y = np.ceil(np.sqrt(x))
np.unique(y, return_counts=True)
```

Powyższa parawidłowość może zostać przedstawiona jako
$$\ceil{\sum^n_{k=1}\sqrt{k}}= \sum^{\ceil{\sqrt{n}}}_{l=0} \sum^{2l+1}_{k=0} l+1 $$

```python
x = np.arange(1,49)
y = np.floor(np.sqrt(x))
fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
fig.set_dpi(300)
ax.scatter(x, y)
ax.set_ylabel(r'$\lfloor\sqrt{k}\rfloor$')
ax.set_xlabel('$k$')
plt.grid()
```
