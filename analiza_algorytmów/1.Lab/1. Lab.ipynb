### 17. Dlaczego $\sum^{2^n}_{k=1} \simeq n \ln 2 \approx \frac{7}{10}n$


```python
n=3
sum=0
for k in range(1, 2**n+1):
    val=1/k
    sum+=val
    print(f"val: {val:.3f} sum: {sum:.3f}")
```
