```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a, b = 0.5, 4.5
```

```python
def func(x):
    return 2*np.sin(x)-np.log(x)
```

```python
x = np.linspace(a, b, 10 ** 4)
y = func(x)


fig, ax = plt.subplots()
ax.plot(x, y, label="Function", color="blue")

# ax.set_aspect('equal')
ax.grid(True, which='both')
fig.set_size_inches(16, 9)
fig.set_dpi(200)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_title(f"Fucntion in [{a},{b}]")
ax.legend()
```

```python
max_iter = 1000
epsilon = 10 ** (-7)

x0=epsilon
x1=b
for i in range(max_iter):
    new_x=(func(x1)*x0-func(x0)*x1)/(func(x1)-func(x0))
    x0=x1
    x1=new_x
    print(f"Iteration {i:<4} | x1={x1:<20} | f(x1)={func(x1)}")
    
    if np.abs(func(x1)) <= epsilon:
        print("func(x)<=epsilon")
        break
```
