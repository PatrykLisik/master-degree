# Metoda Neldera-Meada

## Def simplex
Simplex w $\mathbb{R}^n $ to takie punkty $x_1,x_2,x_3,x_{n+1}$, które nie leżą w jednej (n-1) wymiarowej hiperpłaszczyźnie. 


Założenie: Punkty w simplexie są ponumerowane w taki sposó żeby prawdziwa była 
$$ f:\mathbb{R}^n \to \mathbb{R} $$ 
$$f(x_1) \le f(x_2) \le f(x_3) \le f(x_n) \le f(x_n+1) $$

## Def centroid 
$$ x_0 = \frac{1}{n} \sum^n_{i=1} x_i $$


## Operacja 1 Odbicie(ang. Relexion)
$$ X_r = (1+\alpha ) X_0-\alpha X_{n+1} $$
$$ \alpha > 0 $$
Domyślna wartość $alpha$ to 1

## Operacja 2  -- Ekspansja(ang. Expansion)
$$ X_e = \gamma X_r+(1-\gamma )X_0 $$
$$ \gamma >1 $$
Domyślna wartość $ \gamma $

$$ X_e = 2X_r-X_0 $$



## Operacja 3 -- Kontrakcja (and Contraction)
$$ X_c = \beta X_{n+1} + (1- \beta)X_0 $$
$$ 0<\beta <1 $$
Domyślna wartość $\beta = \frac{1}{2}$



$ X_c = \frac{x_0+x_1+\dots+x_n}{2} $

## Operacja 4 - Redukcja(ang. shrinking)
$$ X_i = \delta(x_i+x_1), dla i=1,2,3,\dots,n,n+1 $$
$$0<\delta<1 $$
Domyślna wartość $\delta = \frac{1}{2}, X_i = \frac{x_i+x_1}{2}, i=1,2,3 \dots , n,n+1 $
