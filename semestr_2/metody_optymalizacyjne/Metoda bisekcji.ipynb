```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X0 = 20
a, b = -5, 1
```

```python
def func(x):
    return -x ** 2 + 5 * x + 10
```

```python
x = np.linspace(a, b, 10 ** 4)
y = func(x)


fig, ax = plt.subplots()
ax.plot(x, y, label="Function", color="blue")

# ax.set_aspect('equal')
ax.grid(True, which='both')
fig.set_size_inches(16, 9)
fig.set_dpi(200)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_title(f"Fucntion in [{a},{b}]")
# labelLines(ax.get_lines(), xvals=(0.1, 1), zorder=2.5)
ax.legend()
```

```python
max_iter = 1000
epsilon = 10 ** (-7)
start = a
end = b
for i in range(max_iter):
    x = (start+end)/2
    if func(a)*func(x)<0:
        end = x
    if func(x)*func(end)<0:
        start = x
    
    print(f"Iteration {i:<4} | x={x:<20} | Solution (a+b)/2={(start+end)/2:<20} | f((a+b)/2)={func((start+end)/2)}")
    if np.abs(func((start+end)/2)) <= epsilon:
        print("func(x)<=epsilon")
        break
    
    if np.abs(x) <= epsilon:
        print("np.abs(x) <= epsilon")
        break
```
