```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def mut_information(p0):
    p1 = 1-p0
    x0y0 = p0*0.531
    x0yd = p0*0.469
    x1y1 = p1*0.531
    x1yd = p1*0.469

    q0 = x0y0
    q1 = x1y1
    qd = 0.469
    hx = -p0*np.log2(p0) - p1*np.log2(p1) 
    HXY = q0*(-x0y0*np.log2(x0y0))+q1*(-x1y1*np.log2(x1y1))+qd*(-x0yd*np.log2(x0yd)-x1yd*np.log2(x1yd))
    return hx - HXY
```

```python
mut_information(0.5)

```

```python
p = np.linspace(0.01,0.99,1000)
```

```python
y = mut_information(p)
```

```python
plt.figure(figsize=(8,5))
plt.plot(p,y)
plt.grid()
plt.savefig("infromacja_wzajemna")
```
